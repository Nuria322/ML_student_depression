```python
# Carga de datos
df = pd.read_csv(r'C:\Users\nuria\OneDrive\Escritorio\ML_student_depression\data\raw\Depression Student Dataset_2.csv')
df.head()
```

### Opciones
- Eliminar la columnas 'City', 'Profession', 'Work Pressure', 'Job Satisfaction', 'Degree'
- Numerar las columnas 'Sleep Duration' y 'Dietary habits'
- Convertir los yes y no en 0 y 1

```python
eliminar = ['City', 'Profession', 'Work Pressure', 'Job Satisfaction', 'Degree']
df = df.drop(columns=eliminar) 
df.head()
```

```python
df.tail(8)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encode = LabelEncoder()
df_categorical =[]
for column in df.columns:
    if df[column].dtypes=="object":
        df_categorical.append(column)
for column in df_categorical: 
    df[column] = encode.fit_transform(df[column])
```

### Transformaciones:
- Gender: Male: 1, Female: 0
- Sleep Duration: 5-6 hours: 0, 7-8 hours: 1, Less than 5 hours: 2, More than 8 hours: 3
- Dietary Habits: Healthy: 0, Moderate: 1, Unhealthy: 3 ¬ø?
