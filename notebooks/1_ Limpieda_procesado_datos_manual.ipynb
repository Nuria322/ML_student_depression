```python
df = pd.read_csv(r'C:\Users\nuria\OneDrive\Escritorio\ML_student_depression\data\raw\Student Depression Dataset.csv')
df.head()
```

### Opciones
- Eliminar la columnas 'City', 'Profession', 'Work Pressure', 'Job Satisfaction', 'Degree'
- Numerar las columnas 'Sleep Duration' y 'Dietary habits'
- Convertir los yes y no en 0 y 1

```python
eliminar = ['id', 'City', 'Profession', 'Work Pressure', 'Job Satisfaction', 'Degree']
df = df.drop(columns=eliminar) 
df.head()
```

```python
df.tail(8)
```

### Transformaciones:
- Gender: Male: 1, Female: 0
- Sleep Duration: 5-6 hours: 0, 7-8 hours: 1, Less than 5 hours: 2, More than 8 hours: 3
- Dietary Habits: Healthy: 0, Moderate: 1, Unhealthy: 3 ¿?

#### Conclusión: 
El LabaelEncoder no atribuye los numeros adecuados, hay que hacerlo manualmente

```python
df.head()
```

```python
gender = {'Male': 1, 'Female':0}
df['Gender'] = df['Gender'].map(gender)
df['Gender'] = df['Gender'].astype(int)
df.head(3)
```

```python
sleep = {'More than 8 hours': 0, '7-8 hours': 1, '5-6 hours': 2, 'Less than 5 hours': 3, 'Others':4}
df['Sleep Duration'] = df['Sleep Duration'].map(sleep)
df['Sleep Duration'] = df['Sleep Duration'].astype(int)
df.head(3)
```

```python
diet = {'Healthy': 0, 'Moderate': 1, 'Unhealthy': 2, 'Others': 3}
df['Dietary Habits'] = df['Dietary Habits'].map(diet)
df['Dietary Habits'] = df['Dietary Habits'].astype(int)
df.head(3)
```

```python
suicidals = {'No': 0, 'Yes': 1}
df['Have you ever had suicidal thoughts ?'] = df['Have you ever had suicidal thoughts ?'].map(suicidals)
df['Have you ever had suicidal thoughts ?'] = df['Have you ever had suicidal thoughts ?'].astype(int)
df.head(3)
```

```python
mentals = {'No': 0, 'Yes': 1}
df['Family History of Mental Illness'] = df['Family History of Mental Illness'].map(mentals)
df['Family History of Mental Illness'] = df['Family History of Mental Illness'].astype(int)
df.head(3)
```

```python
print(df['Dietary Habits'].unique())
```

### Cambios realizados:
- Gender = 'Male': 1, 'Female':0
- Sleep Duration = 'More than 8 hours': 0, '7-8 hours': 1, '5-6 hours': 2, 'Less than 5 hours': 3, 'Others':4
- Dietary Habits = 'Healthy': 0, 'Moderate': 1, 'Unhealthy': 2, 'Others': 3
- Have you ever had suicidal thoughts ? = 'No': 0, 'Yes': 1
- Family History of Mental Illness = 'No': 0, 'Yes': 1
